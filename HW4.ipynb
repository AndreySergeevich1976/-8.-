```python
def transpose_matrix(matrix):
    # Создаем пустую матрицу для транспонированного результата
    transposed = []

    # Проходим по каждому столбцу оригинальной матрицы
    for col in range(len(matrix[0])):
        # Создаем новый ряд для транспонированной матрицы
        new_row = []

        # Проходим по каждой строке оригинальной матрицы
        for row in range(len(matrix)):
            # Добавляем элемент из оригинальной матрицы в новый ряд
            new_row.append(matrix[row][col])

        # Добавляем новый ряд в транспонированную матрицу
        transposed.append(new_row)

    return transposed

# Пример матрицы
original_matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Транспонируем матрицу
transposed_matrix = transpose_matrix(original_matrix)

# Выводим результат
print("Оригинальная матрица:")
for row in original_matrix:
    print(row)

print("Транспонированная матрица:")
for row in transposed_matrix:
    print(row)
```

```python
def create_dict(**kwargs):
    result = {}  # Создаем пустой словарь для результата

    for key, value in kwargs.items():  # Проходим по всем переданным аргументам
        # Пробуем использовать значение в качестве ключа
        try:
            result[value] = key  # Если ключ хешируем, добавляем в словарь
        except TypeError:
            # Если возникла ошибка, используем строковое представление ключа
            result[str(value)] = key

    return result  # Возвращаем полученный словарь

# Пример использования функции
result_dict = create_dict(a=1, b='hello', c=(1, 2), d=[3, 4], e=5.5)

# Выводим результат
print(result_dict)
```

```python
# Константы
PIN = "1456"
PERCENT_PULL = 0.015
OPERATION_ADDED = 3
PERCENT_ADD = 0.03
MIN_PERCENTAGE = 30
MAX_PERCENTAGE = 600
MIN_BANKNOTE = 50
MIN_TAX = 5_000_000
PERCENT_TAX = 0.1

# Лог операций
operation_log = []
operation_history = []  # Лог для операций поступления и снятия

def logging(operation: str):
    """Логирование операций"""
    operation_log.append(operation)

def show_log():
    """Отображение лога операций"""
    print("-- ЛОГ ОПЕРАЦИЙ --")
    for o in operation_log:
        print(o)

def tax_pay(summ: float) -> float:
    """Удержание налога"""
    tax = summ * PERCENT_TAX
    print(f"Налог: {tax:.2f}")
    return summ - tax

def push_cash(balance: float) -> (bool, float):
    """Пополнение счета"""
    summ = float(input("Сумма для пополнения: "))
    if summ % MIN_BANKNOTE == 0:
        if balance > MIN_TAX:
            balance = tax_pay(balance)
        balance += summ
        operation_history.append(f"Пополнение: {summ:.2f}")
        logging(f"ПОПОЛНЕНИЕ - Ok")
        print(f"Баланс пополнен: {summ:.2f}")
        show_balance(balance)
        return True, balance
    else:
        print("Недоступная сумма!")
        logging("ПОПОЛНЕНИЕ - ERROR")
        return False, balance

def pull_cash(balance: float) -> (bool, float):
    """Снятие денег со счета"""
    summ = float(input("Сумма для снятия: "))
    if summ % MIN_BANKNOTE == 0:
        if balance > MIN_TAX:
            balance = tax_pay(balance)
        percent_summ = summ * PERCENT_PULL
        percent_summ = max(MIN_PERCENTAGE, min(percent_summ, MAX_PERCENTAGE))

        if balance - summ - percent_summ < 0:
            print("Недостаточно средств!")
            return False, balance
        else:
            balance -= (summ + percent_summ)
            operation_history.append(f"Снятие: {summ:.2f}, Комиссия: {percent_summ:.2f}")
            logging(f"СНЯТИЕ - Ok")
            print(f"Выдано {summ:.2f}, комиссия {percent_summ:.2f}")
            show_balance(balance)
            return True, balance
    else:
        print("Недоступная сумма!")
        logging("СНЯТИЕ - ERROR")
        return False, balance

def show_balance(summ: float):
    """Отображение баланса"""
    print(f"Баланс счета: {summ:.2f}")

def show_menu(menu: dict) -> int:
    """Отображение меню и получение выбора пользователя"""
    for k, v in menu.items():
        print(f"{k} - {v}")
    result = int(input("> "))
    return result if result in menu.keys() else 0

def apply_interest(balance: float) -> float:
    """Начисление процентов на баланс"""
    interest = balance * PERCENT_ADD
    print(f"Начисление %: {interest:.2f}")
    return balance + interest

def main():
    balance = 0
    operation_counter = 0
    menu_bank = {
        1: "снять",
        2: "пополнить",
        3: "баланс",
        0: "выход",
    }

    authorized = PIN == (input("Введите PIN: "))

    while authorized:
        action = show_menu(menu_bank)
        match action:
            case 1:
                success, balance = pull_cash(balance)
                if success:
                    operation_counter += 1
            case 2:
                success, balance = push_cash(balance)
                if success:
                    operation_counter += 1
            case 3:
                show_balance(balance)
            case _:
                break

        if operation_counter == OPERATION_ADDED:
            operation_counter = 0
            balance = apply_interest(balance)

    else:
        print("Неверный пин-код!")

    show_log()
    print("История операций:")
    for history in operation_history:
        print(history)

if __name__ == "__main__":
    main()
```
