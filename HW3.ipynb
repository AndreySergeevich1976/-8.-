```python
 def find_duplicates(input_list):
    duplicates = set()  # Множество для хранения уникальных дубликатов
    seen = set()  # Множество для хранения всех уже встреченных элементов

    for item in input_list:
        if item in seen:
            duplicates.add(item)  # Если элемент уже встречался, добавляем его в дубликаты
        else:
            seen.add(item)  # Если элемент новый, добавляем его в множество seen

    return list(duplicates)  # Преобразуем множество обратно в список

# Пример использования
input_list = [1, 2, 3, 1, 2, 4, 5, 3]
result = find_duplicates(input_list)
print(result)  # Вывод: [1, 2, 3]
```

```python
 def get_most_common_words(text, n=10):
    # Приводим текст к нижнему регистру и удаляем знаки препинания
    cleaned_text = ''
    for char in text:
        if char.isalnum() or char.isspace():
            cleaned_text += char.lower()

    # Разбиваем текст на слова
    words = cleaned_text.split()

    # Создаем словарь для подсчета слов
    word_counts = {}

    for word in words:
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1

    # Сортируем слова по частоте
    sorted_words = sorted(word_counts.items(), key=lambda item: item[1], reverse=True)

    # Получаем n самых частых слов
    most_common = sorted_words[:n]

    return most_common

# Пример текста (можно заменить на любую статью)
text = """
Python — это язык программирования высокого уровня, интерпретируемый и с динамической типизацией.
Он был создан Гвидо ван Россумом и впервые выпущен в 1991 году.
Python поддерживает несколько программных парадигм, включая объектно-ориентированное, императивное и функциональное программирование.
"""

# Получаем 10 самых частых слов
result = get_most_common_words(text)
print(result)
```

```python
def find_one_combination(items, max_weight):
    # Инициализируем переменные
    current_combination = []
    current_weight = 0

    # Проходим по каждому элементу
    for item, weight in items:
        # Если добавление элемента не превышает максимальный вес
        if current_weight + weight <= max_weight:
            current_combination.append(item)  # Добавляем элемент в комбинацию
            current_weight += weight  # Увеличиваем текущий вес

    return current_combination  # Возвращаем один допустимый вариант

def find_all_combinations(items, max_weight, index=0, current_combination=[], current_weight=0):
    # Если текущий вес превышает максимальный, возвращаем пустой список
    if current_weight > max_weight:
        return []

    # Если достигли конца списка, возвращаем текущую комбинацию
    if index == len(items):
        return [current_combination]

    # Получаем все возможные комбинации без текущего элемента
    without_current = find_all_combinations(items, max_weight, index + 1, current_combination, current_weight)

    # Получаем все возможные комбинации с текущим элементом
    with_current = find_all_combinations(items, max_weight, index + 1, current_combination + [items[index][0]], current_weight + items[index][1])

    # Объединяем оба варианта
    return without_current + with_current

# Словарь с вещами и их массой
gear = {
    'палатка': 2.5,
    'спальный мешок': 1.0,
    'каремат': 0.5,
    'фляга': 0.3,
    'приборы': 0.2,
    'еда': 1.5,
    'фонарик': 0.4,
}

# Преобразуем словарь в список кортежей (название, масса)
items = list(gear.items())

# Максимальная грузоподъемность рюкзака
max_weight = 4.0

# Находим один допустимый вариант комплектации рюкзака
one_combination = find_one_combination(items, max_weight)
print(f"Один допустимый вариант: {one_combination}")

# Находим все возможные варианты комплектации рюкзака
all_combinations = find_all_combinations(items, max_weight)

# Выводим все возможные комплектации
print("Все возможные варианты комплектации рюкзака:")
for i, combination in enumerate(all_combinations):
    print(f"Вариант {i + 1}: {combination}")
```
